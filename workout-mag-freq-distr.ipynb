### Magnitude-Frequency Distribution 

The magnitude frequecy distrbution is given by Gutenberg-Richter (GR) law:

log$_{10}$ [N(>=<i>M$_W$</i>)] = a-b <i>M$_W$</i> &emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp; &emsp; &emsp;(1)

where a and b are a-value and b-value, N is number of events, and M$_W$ refers to moment magnitude <i>Mw</i>.


 


```python
bvalue, avalue, Mmin, Mmax =0.9, 5.0, 4.0, 9.01
mfd.test_sample_GRdistr(bvalue, avalue, Mmin, Mmax, nevents=100000, rseed=20401)
```
